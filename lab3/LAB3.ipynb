```python
import yfinance as yf
import pandas as pd
```

```python
# Just looking at Tesla for now
msft = yf.Ticker("TSLA")

# GETTING ALL THE STOCK INFO
info = msft.info

#Getting just the ticker name
ticker_name = info['symbol']

# Print the ticker name
print(ticker_name)
```

```python
# GETTING THE HISTORICAL DATA OF TESLA STOCK

historical_data = msft.history(period = "1mo")
# historical_data
```

```python
msft.history_metadata
```

```python
# show share count
msft.get_shares_full(start="2022-01-01", end = None)
```

```python
# Show future and historic earnings dates, returns at most next 4 quarters and last 8 quarters by default.
# Note: If more are needed use msft.get_earnings_dates(limit=XX) with increased limit argument.

e_dates = msft.earnings_dates
```

```python
df = yf.download('TSLA', start='2021-01-01', end='2023-01-12')
df['Ticker'] = info['symbol']


# Calculate the daily return and add it as a new column
df['Daily Return'] = df['Close'].pct_change()

n_periods = 20
# Calculate the Simple Moving Average: Moving averages help identify trends
df['SMA'] = df['Close'].rolling(window = n_periods).mean()

# Calculate the Exponential Moving Average: Good with price changes
df['EMA'] = df['Close'].ewm(span = n_periods, adjust = False).mean()
df = df[['Ticker', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Daily Return', 'SMA', 'EMA']]
df
```

```python
# CLEANING THE DF

# used backward fill because the NA's were in the first row
df.fillna(method = 'bfill', inplace = True)
df
```
